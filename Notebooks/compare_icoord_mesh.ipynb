In this notebook we show the difference between in the mesh resulting from different values of the parameter *icoord* (change of coordinates).

```python
import os
import sys
import numpy as np
from pydiva import pydiva2d
import logging
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

# Prepare the input

Indicate the path and names of the 2 mesh files, and define a *figure* directory:

```python
meshdir = '../data/Mesh/icoord/'
meshfile1 = os.path.join(meshdir, 'icoord0_fort.22')
meshtopofile1 = os.path.join(meshdir, 'icoord0_fort.23')
meshfile2 = os.path.join(meshdir, 'icoord1_fort.22')
meshtopofile2 = os.path.join(meshdir, 'icoord1_fort.23')
meshfile3 = os.path.join(meshdir, 'icoord2_fort.22')
meshtopofile3 = os.path.join(meshdir, 'icoord2_fort.23')
figdir = './figures/icoord/'
if not os.path.exists(figdir):
    os.makedirs(figdir)
```

Create a Mesh object and read the information from the files:

```python
mesh01 = pydiva2d.Diva2DMesh()
mesh01.read_from(meshfile1, meshtopofile1)
mesh01.describe()
mesh02 = pydiva2d.Diva2DMesh()
mesh02.read_from(meshfile2, meshtopofile2)
mesh02.describe()
mesh03 = pydiva2d.Diva2DMesh()
mesh03.read_from(meshfile3, meshtopofile3)
mesh03.describe()
```

# Create the figures

```python
fig = plt.figure(figsize=(15, 10))
plt.title("Comparison of mesh with different icoords")

ax = plt.subplot(131)
plt.title("icoord = 0")
mesh01.add_to_plot(linewidth=.1)
ax = plt.subplot(132)
plt.title("icoord = 1")
mesh02.add_to_plot(linewidth=.1)
ax = plt.subplot(133)
plt.title("icoord = 2")
mesh03.add_to_plot(linewidth=.1)
plt.savefig(os.path.join(figdir, "mesh_icoord"), dpi=300, bbox_inches="tight")
plt.show()
plt.close()
```
